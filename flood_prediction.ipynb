```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
test=test.sample(random_state=42,n=450000)
train=train.sample(random_state=42,n=420000)

train.to_csv("resized_train.csv")
test.to_csv("resized_test.csv")
```

```python
train.drop('id',axis=1,inplace=True)
```

```python
scaler=MinMaxScaler(feature_range=(0,1))

X=train.drop(columns='FloodProbability')
Y=train['FloodProbability']

x_train,x_valid,y_train,y_valid=train_test_split(X,Y,test_size=0.2,random_state=42)

x_train=scaler.fit_transform(x_train)
x_valid=scaler.transform(x_valid)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)


pred_trainY=linear_model.predict(x_train)
pred_validY=linear_model.predict(x_valid)
r_train=r2_score(y_train,pred_trainY)
r_valid=r2_score(y_valid,pred_validY)

print(f"R2_score for train: {r_train}")
print(f"R2_score for test: {r_valid}")
```

```python
linear_model.fit(X,Y)

id_column=test['id']
test.drop('id',axis=1,inplace=True)


predict_testY = linear_model.predict(test)


result_df=pd.DataFrame(
    {
        'id':id_column,
        'FloodProbability':predict_testY
        
    }
)

result_df.to_csv('res.csv',index=False)
```
